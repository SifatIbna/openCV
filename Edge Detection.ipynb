```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('test images/noise.png')
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img,threshold1=127,threshold2=127)
```

```python
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img,threshold1=0,threshold2=255)
plt.imshow(edges)
```

```python
med_val = np.median(img)
```

```python
med_val
```

```python
# LOWER THRESHOLD TO EITHER 0 OR 70% OF THE MEDIAN VALUE WHICHEVER IS GREATER
lower = int(max(0,0.7*med_val))
# UPPER THRESHOLD TO EITHER 130% OF THE MEDIAN OR THE MAX 255, WHICHEVER IS SMALLER
upper = int(min(255,1.3*med_val))
```

```python
edges = cv2.Canny(image=img,threshold1=lower,threshold2=upper+100)
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(img,(5,5))
```

```python
edges = cv2.Canny(image=blurred_img,threshold1=lower,threshold2=upper-100)
plt.imshow(edges)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
img = cv2.imread('test images/batman.jpg')
```

```python
plt.imshow(img)
```

```python
 ### MATPLOT color order --> RGB RED GREEN BLUE
 ### OPENCV color order --> BLUE GREEN RED
```

```python
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
img_gray = cv2.imread('test images/batman.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
img_gray.shape ## As the picture is read in GRAYSCALE ,
                ## the shape is not longer 3 dimensional
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python
plt.imshow(fix_img)
```

```python
new_img = cv2.resize(fix_img,(1000,400))
```

```python
plt.imshow(new_img)
```

```python
### Resize by RATIO

w_ratio = 0.8 # 80% of the original width
h_ratio = 0.2 # 20% of the original height

new_img = cv2.resize(fix_img,(0,0),fix_img,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
new_img.shape
```

```python
#fliping
new_img = cv2.flip(fix_img,0) # fliping horizontal
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img,1) # fliping vertical
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img,-1) # fliping horizontal & vertical
plt.imshow(new_img)
```

```python
#writing to a file
```

```python
cv2.imwrite('test images/total_new_pic.jpg',fix_img)
```
